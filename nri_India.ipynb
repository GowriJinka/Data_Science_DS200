```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Tourism_In_India_Statistics_2018-Table_2.1.2.csv')
```

```python
df = data.iloc[:-1,:]
```

```python
years=['2015','2016','2017']
for year in years:
    plt.bar(df['Months'], df[year], width = 0.6, align = 'center')
    plt.xticks(rotation=45)
    plt.xlabel('Month')
    plt.ylabel('Tourist Numbers')
    plt.title('Tourist Numbers Over Months ({})'.format(year))
    plt.show()
```

```python

# Set the positions and width for the bars
n = len(df)
x = np.arange(n)
width = 0.2

# Create the bar graph
plt.figure(figsize=(12, 6))
for year in ['2015', '2016', '2017']:
    plt.bar(x, df[year], width, label=year)
    x = x + width

# Customize the plot
plt.xlabel('Months')
plt.ylabel('Number of Tourists')
plt.title('Number of Tourists in India by Month (2015-2017)')
plt.xticks(np.arange(n), df['Months'])
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
sns.set(style="whitegrid")

# Line Plot: Trends in tourist numbers over the months for each year
plt.figure(figsize=(12, 6))
sns.lineplot(x='Months', y='2015', data=df, label='2015')
sns.lineplot(x='Months', y='2016', data=df, label='2016')
sns.lineplot(x='Months', y='2017', data=df, label='2017')
plt.xticks(rotation=45)
plt.xlabel('Month')
plt.ylabel('Tourist Numbers')
plt.title('Tourist Numbers Over Months (2015-2017)')
plt.legend()
plt.show()
```

```python
months = df['Months']

# Create line plots for percentage growth
plt.figure(figsize=(15, 10))

plt.plot(months, df['Percentage Growth - 2016/2015'], marker='o', label='2016/2015 Growth', linestyle='-', color='b', markersize=8)
plt.plot(months, df['Percentage Growth - 2017/2016'], marker='s', label='2017/2016 Growth', linestyle='--', color='g', markersize=8)

# Customize the plot
plt.xlabel('Month')
plt.ylabel('Percentage Growth (%)')
plt.title('Percentage Growth of Tourists in India (2015-2017)')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Heatmap: Percentage growth between 2015/2016 and 2016/2017
growth_data = df[['Percentage Growth - 2016/2015', 'Percentage Growth - 2017/2016']]
plt.figure(figsize=(8, 6))
sns.heatmap(growth_data, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.xlabel('Year')
plt.ylabel('Month')
plt.title('Percentage Growth Between 2015/2016 and 2016/2017')
plt.show()
```
